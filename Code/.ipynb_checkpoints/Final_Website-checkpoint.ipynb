```python
app = dash.Dash(__name__)

# ------------------------------------------------------------------------------
# Import and clean data (importing csv into pandas)
df = pd.read_csv("districts.csv")

#creating df for each state for choropleth plot
def generate_states(df, state):
    india_states = json.load(open(state.lower()+"_district.json", "r"))

    state_id_map = {}
    for feature in india_states["features"]:
        feature["id"] = feature["properties"]["dt_cen_cd"]
        state_id_map[feature["properties"]["district"]] = feature["id"]
    
    if(state == 'Tamil Nadu'):
        df.loc[df.District == "Tenkasi", "District"] = "Tirunelveli"
        df.loc[df.District == "Ranipet", "District"] = "Vellore"
        df.loc[df.District == "Tirupathur", "District"] = "Vellore"
        df.loc[df.District == "Chengalpattu", "District"] = "Kancheepuram"
        df.loc[df.District == "Kallakurichi", "District"] = "Viluppuram"
    
    df5 = df.copy()
    df5 = df5.loc[df["State"] == state]
    df5 = df5.sort_values(by="Date")
    df5 = df5.drop_duplicates(subset = ['District'],keep = 'last')
    df5 =df5.drop(['Tested', 'Other'], axis= 1)
    df5 = df5.reset_index(drop=True)

    geo_df = pd.DataFrame(data=[st['properties']['district'] for st in india_states['features']], columns=['State Names in GeoJSON'])
    data_df = pd.DataFrame(df5["District"].unique(), columns=['State Names in DATA'])
    compare = geo_df.merge(data_df,
                 how='outer',
                 left_on='State Names in GeoJSON',
                 right_on='State Names in DATA')

    compare[["State Names in GeoJSON", "State Names in DATA"]] = compare[["State Names in GeoJSON", "State Names in DATA"]].fillna("")
    df5 = df5.loc[~((df5["District"] == "Other State") | (df5["District"] == "Unknown") | (df5['District'] == 'Foreign Evacuees') | 
                    (df5["District"] == "Railway Quarantine") | (df5["District"] == "Airport Quarantine"))] 
    df5 = df5.reset_index(drop = True)

    df5['id'] = df5["District"].apply(lambda x: state_id_map[x])
    print(df5.shape)
    return df5, india_states
```

```python
@app.callback(
#     [Output(component_id='output_container', component_property='children')
     Output(component_id='my_bee_map', component_property='figure'),
    [Input(component_id='slct_year', component_property='value'),
     Input(component_id='radio', component_property='value'),
     Input(component_id='check', component_property='value')]
)
def update_graph(option_slctd, option, option_list):
    
        
    if(option_slctd == 'India'):
        india_states = json.load(open("states_india_copy.json", "r"))
        state_id_map = {}
        for feature in india_states["features"]:
            feature["id"] = feature["properties"]["state_code"]
            state_id_map[feature["properties"]["st_nm"]] = feature["id"]
        
        plot_df['id'] = plot_df["State"].apply(lambda x: state_id_map[x])
        
        color = ''
        if(option == 'Confirmed'):
            color = px.colors.sequential.Blues
        elif(option == 'Recovered'):
            color = px.colors.sequential.Greens
        else:
            color = px.colors.sequential.YlOrRd
        fig = px.choropleth(
        plot_df,
        locations="id",
        geojson=india_states,
        color=option,
        hover_name="State",
        hover_data=option_list,
        color_continuous_scale= color,
        title=option_slctd,
        template='plotly_dark'
        )
        fig.update_geos(fitbounds="locations", visible=False)
        fig.update_layout(height=750, title_x = 0.44)
        
        return fig
    
    print(option_slctd)
    print(type(option_slctd))
    
    print(option)
    print(type(option))
    #hov = option
    

    dff = generate_states(df,option_slctd)
    #dff = dff[dff["State"] == option_slctd]
    
    if(option == "Confirmed"):
        fig = px.choropleth(
        dff[0],
        locations="id",
        geojson=dff[1],
        color=option,
        hover_name="District",
        hover_data= option_list,
        color_continuous_scale=px.colors.sequential.Blues,
        title=option_slctd,
        template='plotly_dark',
        #width = 10,
        #height = 10
        )
        fig.update_geos(fitbounds="locations", visible=False)
        fig.update_layout(height=750, title_x = 0.44)
        return fig

    #color = px.color.sequential()
    if(option == 'Recovered'):
        fig = px.choropleth(
        dff[0],
        locations="id",
        geojson=dff[1],
        color=option,
        hover_name="District",
        hover_data= option_list,
        color_continuous_scale=px.colors.sequential.Greens,
        title=option_slctd,
        template='plotly_dark',
        #width = 10,
        #height = 10
        )
        fig.update_geos(fitbounds="locations", visible=False)
        fig.update_layout(height=750, title_x = 0.44)
        return fig
    
    if(option == 'Deceased'):
        fig = px.choropleth(
        dff[0],
        locations="id",
        geojson=dff[1],
        color=option,
        hover_name="District",
        hover_data= option_list,
        color_continuous_scale=px.colors.sequential.YlOrRd,
        title=option_slctd,
        template='plotly_dark',
        #width = 10,
        #height = 10
        )
        fig.update_geos(fitbounds="locations", visible=False)
        fig.update_layout(height=750, title_x = 0.44)
        return fig

    
@app.callback(
    Output(component_id='line_chart', component_property='figure'),
    [Input(component_id='drop_cases', component_property='value'),
     Input(component_id='states', component_property='value')]
)
def update_line_chart(dropdown, states_select):
    if(states_select == 'India'):
        fig1 = px.line(data_frame = df1, x="Date", y=dropdown, template = 'plotly_dark')
    
    else:
#         df2 = df2.loc[df2['State'] == states_select]
        fig1 = px.line(data_frame = df2.loc[df2['State'] == states_select], x="Date", y=dropdown, template = 'plotly_dark')
    
    return fig1

@app.callback(
    [Output(component_id = 'acc_pred', component_property = 'figure'),
    Output('table1', 'children')],
    [Input('drop_models', 'value')
    ,Input('dist_slct', 'value')]
)
def df_acc_pred(models_drop, select_dist):
    
    dfgf = pd.read_csv(models_drop.lower() + "_" + select_dist.lower() + ".csv")
        #+ "_" + select_dist.lower()

    #dfgf = pd.DataFrame(dfgf)
    
    fig2 = px.line(data_frame = dfgf, x="Date", y=dfgf.columns[2:4], template = 'plotly_dark')

    return fig2, html.Div([dash_table.DataTable(
            data= dfgf.to_dict('records'),
            columns=[{"name": i, "id": i} for i in dfgf.columns],
            page_action='none',
            fixed_rows={'headers': True},
            style_table={'height': '300px', 'width': '100%', 'overflowY': 'auto'},
            style_header={'backgroundColor': 'rgb(30, 30, 30)'},
            style_cell={
                'minWidth': '180px', 'width': '100px', 'maxWidth': '180px',
                'backgroundColor': 'rgb(50, 50, 50)',
                'color': 'white'
            }
        )]
                         )
```
