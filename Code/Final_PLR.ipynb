```python
def daily(sorted_data,district):
    temp = sorted_data.loc[sorted_data["District"] == district[0]].copy()
#     temp['Confirmed'] = temp['Confirmed'].diff()
    temp['Confirmed'] = temp['Confirmed'].fillna(0)
    temp = temp.astype({"Confirmed": int})
    for i in range(1,len(district)):
        temp1 = sorted_data.loc[sorted_data["District"] == district[i]].copy()
#         temp1['Confirmed'] = temp1['Confirmed'].diff()
        temp1['Confirmed'] = temp1['Confirmed'].fillna(0)
        temp1 = temp1.astype({"Confirmed": int})
        temp = temp.append(temp1)
    return temp
```

* First quarter    : Dated between 01/01/2020 and 31/03/2020

* Third quarter  : Dated between 01/07/2020 and 30/09/2020
* Fourth quarter : Dated between 01/10/2020 and 31/12/2020



```python
#Function to generate dataframe based on start and end dates 
def quarter_generator(df, start_date:str ,end_date:str):

    # To convert the input taken from user into datetime objects
    
    start_date1 = pd.to_datetime(start_date, format = '%d-%m-%Y', errors = 'ignore')  #start_date1 is the datetime object for start_date
    end_date1 = pd.to_datetime(end_date, format = '%d-%m-%Y', errors = 'ignore')      #end_date1 is the datetime object for end_date
    
    
    # To obtain the quarter based on start_date1 and end_date1
    
    quarter = df.loc[df['Date'] <= end_date1]
    quarter = quarter.loc[quarter['Date'] >= start_date1]  
    return quarter
```

```python

#Plot train data
def plot_df_train(x,y,xHat,yHat):
     #Plotting the values
    plt.scatter(x, y, color = 'blue')  
    plt.title("Piecewise Linear Regression Best Fit Training data") 
    plt.xlabel('Date') 
    plt.ylabel('Confirmed Cases') 
    plt.plot(xHat, yHat, color = 'red')

#Plot test data    
def plot_df_test(x,y,xHat,yHat):
     #Plotting the values
    plt.plot(x, y, color = 'blue');
    plt.plot(xHat, yHat, color = 'red');
    #plt.scatter(, y) 
    #sns.set(rc={'figure.figsize':(15,10)})
    #sns.scatterplot(x, y)
    plt.title("Actual vs Predicted values for Confirmed Cases") 
    plt.xlabel('Date') 
    plt.ylabel('Confirmed Cases') 
    plt.show()
    
#Piecewise Linear Regression
def districts_model(df_train, n, df_test):

    #Loading testing data
    q1_2021 = df_test['Date'].apply(lambda x : x.toordinal()).to_list()
    q1_testx = np.linspace(min(q1_2021), max(q1_2021), num = len(q1_2021))

    #print(q1_2021)
    #Convering datetime to ordinal 
    try:
        df_train['Date'] = df_train['Date'].map(dt.datetime.toordinal)
    except:
        pass
    
    #Loading train data
    x = np.array(df_train['Date'])
    y = np.array(df_train['Confirmed'])

    #Initialize piecewise linear fit with x and y data
    myPWLF = pwlf.PiecewiseLinFit(x,y)
    
    #Fit the data for n line segments
    res = myPWLF.fit(n)
    
    #Predict for the determined points
    xHat = np.linspace(min(x), max(x), num = len(x))
    yHat = myPWLF.predict(xHat)
    
    #Predicting for the 1st quarter of 2021
    q1_testy = myPWLF.predict(q1_2021)
    
    #print(type(q1_testy))
    #df_train['Date'] = df_train['Date'].map(pd.Timestamp.fromordinal)
    
    xdate = np.array(df_train['Date'])
    ydate = np.array(df_train['Confirmed'])
    
    #Plot train data
    #plot_df_train(xdate,ydate,xHat,yHat)
    
    #Creating new dataframe for actual and predicted values
    predicted_df = pd.DataFrame(columns = ['Date', 'Actual Confirmed', 'Predicted Confirmed', 'Absolute Error']) 

    #Extracting values according to position
    predicted_df["Date"] = df_test.iloc[:,0]
    predicted_df["Actual Confirmed"] = df_test.iloc[:,3]
    predicted_df["Predicted Confirmed"] = pd.Series(q1_testy.astype(int), index=predicted_df.index)
    mn = predicted_df['Predicted Confirmed'].mean()
    std = predicted_df['Predicted Confirmed'].std()
    predicted_df['Min Predicted Confirmed'] = (predicted_df['Predicted Confirmed'] - 2 * std).map(int)
    predicted_df['Max Predicted Confirmed'] = (predicted_df['Predicted Confirmed'] + 2 * std).map(int)
    
    predicted_df["Absolute Error"] = (predicted_df['Actual Confirmed'] - predicted_df['Predicted Confirmed']).map(abs)
    predicted_df["Accuracy"] = (1 - (predicted_df['Absolute Error']/predicted_df['Actual Confirmed'] ))*100
    predicted_df['% of incorrect prediction'] = (predicted_df['Absolute Error']/predicted_df['Actual Confirmed'])*100
    
    #r2_train = sklearn.metrics.r2_score(y, yHat)
    
    return predicted_df
```

```python
for dist in dist_d.keys():
    seg = 4
    predicted_max = 10000000
    for train in dist_d[dist]:
        if (train != "best"):
            predicted = districts_model(dist_d[dist][train]['df'], seg,\
                                        quarter2_21.loc[quarter2_21['District'] == dist])
            if(predicted[:n]['Absolute Error'].mean() < predicted_max):
                dist_d[dist]["best"] = predicted[:n].reset_index(drop = True)
                predicted_max = predicted[:n]['Absolute Error'].mean()
    dist_d[dist]['best'].to_csv("plr"+dist[:3]+".csv")
```
